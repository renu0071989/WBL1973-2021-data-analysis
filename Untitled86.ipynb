```python
df['CalendarYear'] = pd.to_datetime(df['CalendarYear'], format='%Y')

# Independence years dictionary
independence_years = {
    'Eritrea': 1993,
    'South Sudan': 2011,
    'Timor Leste': 2012
}

# Create a list to store the index values of rows to be dropped
rows_to_drop = []

# Filter the rows based on the independence years and add their index values to the list
for country, year in independence_years.items():
    rows_to_drop.extend(df.loc[(df['Economy'] == country) & (df['CalendarYear'] < pd.to_datetime(year, format='%Y'))].index)

# Drop the rows from the original DataFrame 'df'
df = df.drop(rows_to_drop)
```

```python
df.columns
```

```python

# Create a dictionary to store the reform counter for each column
reform_counter = {col: 0 for col in df.columns if "?" in col}

# Create a new DataFrame containing the previous row values for each column
previous_row_df = df.shift()

# Iterate over each row
for idx, row in df.iterrows():
    # Reset the reform counter for each column
    for col in reform_counter:
        reform_counter[col] = 0

    # Skip the first row since it doesn't have a previous row
    if idx == 0:
        continue
    
    # Iterate over each column and check for "?" in the column name
    for col, val in row.iteritems():
        if "?" in col:
            # Check if the current row's value is "No" and the previous row's value is "Yes"
            if val == "No" and previous_row_df.loc[idx, col] == "Yes":
                # Get the name of the next column (to check adjacent column values)
                next_col = df.columns[df.columns.get_loc(col) + 1] if df.columns.get_loc(col) < len(df.columns) - 1 else None
                if next_col is not None:
                    # Check if the adjacent column values also change
                    if val != previous_row_df.loc[idx, next_col]:
                        reform_counter[col] += 1  # Increment the reform counter
                else:
                    reform_counter[col] += 1  # Increment the reform counter (if the last column has "Yes" and "No" changes)

    # Add the reform counter values as a new column to the DataFrame
    df.at[idx, 'Reform_Count'] = sum(reform_counter.values())

# Save the DataFrame to an Excel file
df.to_excel('reform_data.xlsx', index=False)



```

```python
import pandas as pd
import re



column_groups = {
    'mobility': ['Legal Basis_mobility', 'Legal Basis.1_mobility', 'Legal Basis.2_mobility', 'Legal Basis.3_mobility'],
    'workplace': ['Legal Basis_workplace', 'Legal Basis.1_workplace', 'Legal Basis.2_workplace', 'Legal Basis - Criminal Penalties', 'Legal Basis - Civil Remedies'],
    'pay': ['Legal Basis_pay', 'Legal Basis.1_pay', 'Legal Basis.2_pay', 'Legal Basis.3_pay'],
    'marriage': ['Legal Basis_marriage', 'Legal Basis.1_marriage', 'Legal Basis.2_marriage', 'Legal Basis.3_marriage', 'Legal Basis.4_marriage'],
    'parenthood': ['Legal Basis_parenthood', 'Legal Basis.1_parenthood', 'Legal Basis.2_parenthood', 'Legal Basis.3_parenthood', 'Legal Basis.4_parenthood'],
    'entrepreneurship': ['Legal Basis_enter', 'Legal Basis.1_enter', 'Legal Basis.2_enter', 'Legal Basis.3_enter'],
    'assets': ['Legal Basis_assets', 'Legal Basis.1_assets', 'Legal Basis.2_assets', 'Legal Basis.3_assets', 'Legal Basis.4_assets'],
    'pension': ['Legal Basis_pension', 'Legal Basis.1_pension', 'Legal Basis.2_pension', 'Legal Basis.3_pension']
}

# Function to clean legal basis column
def clean_legal_bases(column_values):
    # Combine all column values into a single string
    merged_bases = ";".join(column_values)
    
    # Remove unwanted strings using regex
    merged_bases = re.sub(r'No applicable provisions could be located|No restrictions could be located|No\s+---', '', merged_bases)
    
    # Remove leading and trailing spaces
    merged_bases = merged_bases.strip()
    
    # Remove extra spaces in between
    merged_bases = re.sub(r'\s+', ' ', merged_bases)
    
    # Remove ending period
    if merged_bases.endswith('.'):
        merged_bases = merged_bases[:-1]
    
    # Remove duplicates by splitting, using set, and rejoining
    merged_bases = ";".join(sorted(set(merged_bases.split(";"))))
    
    # Remove semicolon in the beginning of each row
    if merged_bases.startswith(';'):
        merged_bases = merged_bases[1:]
    
    return merged_bases

# Create a new DataFrame to store cleaned legal basis for each group
cleaned_legal_basis_df = pd.DataFrame()

# Loop through each group and clean the legal basis columns
for group, columns in column_groups.items():
    cleaned_legal_basis_df[group] = df_cleaned[columns].apply(clean_legal_bases, axis=1)

# Merge "Economy", "Income Group", and "Report Year" columns
cleaned_legal_basis_df.insert(0, 'Economy', df_cleaned['Economy'])
cleaned_legal_basis_df.insert(1, 'Income Group', df_cleaned['Income Group'])
cleaned_legal_basis_df.insert(2, 'CalendarYear', df_cleaned['CalendarYear'])

# Print the cleaned DataFrame with merged legal bases for each group
cleaned_legal_basis_df
```

```python
output_file = 'output_file.xlsx'

# Save the DataFrame to an Excel file
cleaned_legal_basis_df.to_excel(output_file, index=False)

print(f"DataFrame saved to {output_file} successfully.")
```
